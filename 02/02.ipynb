```python
with open('input') as f:
    data = [[int(d) for d in line.strip().split('x')] for line in f]
data
```

```python
import math

def calc_paper(l, w, h):
    return 2*l*w + 2*w*h + 2*h*l + math.prod(sorted([l, w, h])[:2])
```

```python
print("Part 1:")
print(sum(calc_paper(l, w, h) for l, w, h in data))
```

```python
def calc_ribbon(l, w, h):
    return 2*sum(sorted([l, w, h])[:2]) + math.prod([l, w, h])
```

```python
print("Part 2:")
print(sum(calc_ribbon(l, w, h) for l, w, h in data))
```
